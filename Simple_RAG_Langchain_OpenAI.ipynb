```python
import os
from google.colab import userdata
```

# Initialize openAI LLM

```python
from langchain_openai import ChatOpenAI

os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')

llm = ChatOpenAI(
    model = 'gpt-3.5-turbo',
    temperature=0
  )
```

# Initialize the embedding model

```python
# using openAi embedding models
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")  # embedding model
```

# Loading some documents

```python
# here i will give some documents. not pdf files or web sites. just some texts.
# this is for only we giving texts.
from langchain_core.documents import Document

documents = [
    Document(
        page_content="India secured their second ICC Men's T20 World Cup title by defeating South Africa in a thrilling final held on June 29, 2024, at Kensington Oval in Bridgetown, Barbados. ",
        metadata={"source":"cricket news"},
    ),
    Document(
        page_content=" The Premier League has several key matches scheduled for April and May 2025. Notably, the Merseyside derby between Liverpool and Everton is set for April 2, while the Manchester derby featuring Manchester United and Manchester City is slated for April 6. These fixtures are crucial as teams vie for top positions in the league.",
        metadata={"source":"football news"},
    ),
    Document(
        page_content="​In the 2024 Sri Lankan presidential election held on September 21, Anura Kumara Dissanayake of the National People's Power (NPP) emerged victorious. The election proceeded to a second round of counting, where Dissanayake secured 55.89% of the votes, defeating Sajith Premadasa of the Samagi Jana Balawegaya (SJB), who received 44.11%. This marked the first time in Sri Lanka's history that a presidential election required a second round of counting.",
        metadata={"source":"election news"},
    ),
    Document(
        page_content="Agentic AI represents a significant evolution in artificial intelligence, characterized by systems that can autonomously understand, plan, and execute goals. Unlike traditional generative AI, these agents are designed to adapt in real-time, integrating data from diverse sources to perform complex tasks. Practical applications span optimizing supply chains, enhancing customer support, and managing IT infrastructures, marking a transformative shift in how AI interacts with and utilizes data.",
        metadata={"source":"AI news"}
    )
            ]
```

# save the above data in the vector space.

```python
from langchain_chroma import Chroma

vectorstore = Chroma.from_documents(
    documents,   # give our documents
    embedding = embedding_model  # our embedding model
  )

# after that we have completely stored our text data in our vector db
```

# Similarity search

```python
results = vectorstore.similarity_search("match")   # enter the texts that you need to search

for result in results:
  print("----------------")
  print(result.page_content)
  print(result.metadata)
```

```python
results = vectorstore.similarity_search("Deep Learning")   # enter the texts that you need to search

for result in results:
  print("----------------")
  print(result.page_content)
  print(result.metadata)

# these results comes with comparing the data in our created vectorspace database.
```

```python
# to see the emmbedding of a query

query_embedding = embedding_model.embed_query("Deep Learning")
print(query_embedding)
len(query_embedding)

# this shows the numerical values for the word "Deep Learning"
```

```python
# similarity search using query embedding

results = vectorstore.similarity_search_by_vector(query_embedding)

for result in results:
  print("----------------")
  print(result.page_content)
  print(result.metadata)

# it also outputs the correct news
```

# create retriever for retrieve data from vector space

```python
# make the vector object as retriever
# retrive data from vector space using retriever object

retriever = vectorstore.as_retriever(
    search_type = "similarity", # method to compare user input and vector space embeddings
    search_kwargs = {"k":1}   # number of output results = 1
    # if we give k=2, most importent 2 documents will be outputted.
)

# take output velues through retriever object
batch_results = retriever.batch(["Deep Learning", "Cricket"]) # give text as list(text list)
# if we give one text, it give one output. for 2 texts, it gave 2 outputs.

for result in batch_results:
  print("----------------")
  for doc in result:
    print(doc.page_content)
    print(doc.metadata)
```

# Connect everything and make the prompt

```python
# prompt template
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough

message = """
Answer this question using provided context only

{question}

Context:
{context}
"""
# telling the language model, check the context and give a result for the asking question

# make a chat prompt template
# give the above full message as human message
# it need to output an AI message
prompt = ChatPromptTemplate.from_messages([("human", message)])
```

# Use the LLM using the prompt

```python
chain = {"context": retriever, "question": RunnablePassthrough()} | prompt | llm

# context = retriever response
# here question = RunnablePassthrough()
```

```python
response = chain.invoke("how are you?")
print(response.content)

# when we call using chain, "what is deep learning?" is the runnable. (the passing text when calling is the runnable)
```
